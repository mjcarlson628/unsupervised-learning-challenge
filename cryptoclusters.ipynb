```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
crypto_data = pd.read_csv("crypto_data.csv")
print(len(crypto_data))
```

```python
crypto_data.head()
```

```python
# remove rows that are not being traded
df = crypto_data[crypto_data["IsTrading"] == True].drop(columns = "IsTrading")

# remove rows with any NaN values
df = df.dropna()

# remove currencies with 0 total coins mined
df = df[df["TotalCoinsMined"] > 0]

# remove the coin names
coins_df = df["CoinName"]
df = df.drop(columns = ["Unnamed: 0", "CoinName"])

# convert totalcoinsupply from string to float
df["TotalCoinSupply"] = df["TotalCoinSupply"].astype(float)
```

```python
# convert algorithm and prooftype to numerical data
dums = pd.get_dummies(df)
dums
```

```python
# scale data using standard and minmax scalers
crypto_scaled_m = MinMaxScaler().fit_transform(dums)
crypto_scaled_s = StandardScaler().fit_transform(dums)
```

```python
# reduce dimensions using PCA
pca_m = PCA(n_components = 0.9).fit_transform(crypto_scaled_m)
pca_s = PCA(n_components = 0.9).fit_transform(crypto_scaled_s)
```

```python
df_pca_m = pd.DataFrame(pca_m)
df_pca_m.head()
```

```python
# PCA results in a much greater number of factors using Standard Scaler
df_pca_s = pd.DataFrame(pca_s)
df_pca_s.head()
```

```python
# TSNE learning rate samples
lr = [10, 50, 100, 200, 500, 1000]

for z in lr:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 4))
    fig.suptitle(f"Learning Rate: {z}")
    
    tsne_m = TSNE(learning_rate = z, random_state = 1).fit_transform(df_pca_m)
    ax1.scatter(tsne_m[:,0], tsne_m[:,1], c = "blue")
    ax1.set_title("MinMaxScaler")
    
    tsne_s = TSNE(learning_rate = z, random_state = 1).fit_transform(df_pca_s)
    ax2.scatter(tsne_s[:,0], tsne_s[:,1], c = "red")
    ax2.set_title("StandardScaler")
    
    plt.show()
```

```python
# Find the "best" number of clusters using KMeans
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=1)
    km.fit(df_pca_m)
    inertia.append(km.inertia_)

elbow_data_m = {"k": k, "inertia": inertia}
df_elbow_m = pd.DataFrame(elbow_data_m)

inertia = []
for i in k:
    km = KMeans(n_clusters=i, random_state=1)
    km.fit(df_pca_s)
    inertia.append(km.inertia_)

elbow_data_s = {"k": k, "inertia": inertia}
df_elbow_s = pd.DataFrame(elbow_data_s)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 4))
ax1.plot(df_elbow_m['k'], df_elbow_m['inertia'], "o-", c = "blue")
ax1.set_title("KMeans Elbow Plot - MinMaxScaler")

ax2.plot(df_elbow_s['k'], df_elbow_s['inertia'], "o-", c = "red")
ax2.set_title("KMeans Elbow Plot - StandardScaler")

for p in [ax1, ax2]:
    p.set_xticks(range(len(k) + 1))
    p.set_xlabel('Number of clusters')
    p.set_xlim(0, len(k) + 1)
    p.set_ylabel('Inertia')
    
plt.show()
```

```python
# color code based on scaling method and potential elbow points
tsne_m = TSNE(learning_rate = 200, random_state = 1).fit_transform(df_pca_m)
tsne_s = TSNE(learning_rate = 200, random_state = 1).fit_transform(df_pca_s)
```

```python
cl = [3, 5, 7, 9]
for c in cl:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8, 4))
    fig.suptitle(f"Number of Clusters: {c}")
    
    predict_m = pd.DataFrame(KMeans(n_clusters=c, random_state=1).fit(df_pca_m).predict(df_pca_m))
    ax1.scatter(tsne_m[:,0], tsne_m[:,1], c = predict_m, cmap = "gist_rainbow")
    ax1.set_title("MinMaxScaler")
    
    predict_s = pd.DataFrame(KMeans(n_clusters=c, random_state=1).fit(df_pca_s).predict(df_pca_s))
    ax2.scatter(tsne_s[:,0], tsne_s[:,1], c = predict_s, cmap = "gist_rainbow")
    ax2.set_title("StandardScaler")
    
    plt.show()
```

This list of cryptocurrencies can be clustered together. Based on the number of coins of each currency relative to the minimum and maximum numbers of coins of all currencies, 5 and 7 would be good choices for the number of clusters.
